```python
from typing import List
from sklearn.metrics import r2_score

# validation set

print('\nVALIDATION SET')
indices_val = experiment['indices/val']
graphs_val = [index_data_map[index]['metadata']['graph'] for index in indices_val]
print(' * num validation graphs: ', len(graphs_val))

results_val: List[dict] = model.forward_graphs(graphs_val)
out_true_val = np.concatenate([graph['graph_labels'] for graph in graphs_val])
out_pred_val = np.concatenate([result['graph_output'] for result in results_val])
errors_val = np.abs(out_true_val - out_pred_val)

r2_val = r2_score(out_true_val, out_pred_val)
print(' * r2 score:', r2_val)

# test set

print('\nTEST SET')
indices_test = experiment['indices/test']
graphs_test = [index_data_map[index]['metadata']['graph'] for index in indices_test]
print(' * num test graphs: ', len(graphs_test))

results_test = model.forward_graphs(graphs_test)
out_true_test = np.concatenate([graph['graph_labels'] for graph in graphs_test])
out_pred_test = np.concatenate([result['graph_output'] for result in results_test])
errors_test = np.abs(out_true_test - out_pred_test)

r2_test = r2_score(out_true_test, out_pred_test)
print(' * r2 score:', r2_test)
```

```python
from typing import List
import torch
from torch_geometric.nn import GINConv, global_mean_pool
from torch_geometric.data import DataLoader
import torch.nn as nn

# Define a function to calculate the NLL loss
def nll_loss(mean, var, target):
    """
    Calculates the negative log-likelihood loss.

    Args:
        mean (torch.Tensor): Predicted mean values.
        var (torch.Tensor): Predicted variance values.
        target (torch.Tensor): True target values.

    Returns:
        torch.Tensor: The NLL loss.
    """
    # Ensure variance is non-negative
    #sigma = torch.clamp(sigma, min=1e-8)  # Adding a small constant to avoid division by zero
    
    std = torch.sqrt(var)
    # Calculate the Gaussian NLL loss
    nll = 0.5 * std.detach() * (torch.log(2 * 3.14 * var) + ((target - mean).pow(2) / var))
    loss = torch.mean(nll)

    return loss

def calibrate_ensemble(model: MeganEnsemble,
                       graphs: List[dict],
                       epochs: int = 10_000,
                       learning_rate: float = 0.01,
                       ) -> np.ndarray:
    
    num_models: int = len(model.models)
    factors = torch.ones(1, requires_grad=True)
    
    # targets: (num_graphs, )
    targets = torch.tensor([graph['graph_labels'] for graph in graphs])
    
    results = model.forward_graphs(graphs)
    # predictions: (num_graphs, num_models)
    predictions = torch.tensor(np.stack([result['graph_output_stack'].squeeze() for result in results]))
    
    optimizer = torch.optim.Adam([factors], lr=learning_rate)
    for epoch in range(epochs):
        
        optimizer.zero_grad()
        
        predictions_mean_ = torch.mean(predictions, dim=1, keepdim=True)
        predictions_new = predictions_mean_ + factors.unsqueeze(0) * (predictions - predictions_mean_)
        
        predictions_mean = predictions_new.mean(dim=1)
        predictions_var = predictions_new.var(dim=1)
        
        loss = nll_loss(predictions_mean, predictions_var, targets)
        loss.backward()
        optimizer.step()
        
        if epoch % 100 == 0:
            print(f'epoch {epoch}/{epochs} - loss: {loss.detach().numpy()}')
        
    return factors.detach().numpy()

factors = calibrate_ensemble(
    model=model,
    graphs=graphs_val,
    epochs=10_000,
)

print('factors')
pprint(factors)
    
```

```python
from scipy.stats import pearsonr, spearmanr
from graph_attention_student.metrics import rll_score

print('applying calibration factors...')
member_predictions_val = np.concatenate([result['graph_output_stack'] for result in results_val])
mean_prediction_val = np.mean(member_predictions_val, axis=1, keepdims=True)
member_predictions_val = mean_prediction_val + factors * (member_predictions_val - mean_prediction_val)

mean_prediction_val = np.mean(member_predictions_val, axis=1)
std_prediction_val = np.std(member_predictions_val, axis=1)
errors_val = np.abs(out_true_val - mean_prediction_val)

r2_val = r2_score(out_true_val, mean_prediction_val)
print(' * r2 score:', r2_val)

# Calculate Pearson and Spearman correlation
pearson_corr, _ = pearsonr(std_prediction_val, errors_val)
spearman_corr, _ = spearmanr(std_prediction_val, errors_val)
rll_value = rll_score(out_true_val, mean_prediction_val, std_prediction_val)

# Plotting
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.scatter(std_prediction_val, errors_val, alpha=0.5, color='skyblue', linewidths=0.0)
plt.xlabel('Predicted Standard Deviation')
plt.ylabel('Absolute Error')
plt.title(f'Pearson: {pearson_corr:.2f}, Spearman: {spearman_corr:.2f}\n'
          f'RLL: {rll_value:.2f}')
plt.grid(True, alpha=0.2)
plt.show()
```
